```python
ising = Ising(n=32)
results = ising.run_multi_sim()
temperatures_32, magnetizations_32, energies_32, heat_capacities_32 = results['temperatures'], results['magnetizations'], results['energies'], results['heat_capacities']
```

```python
x = temperatures_32

fig, ax = plt.subplots(figsize=(12, 10))
ax.axvline(x=2.269, color='red', linestyle='--', linewidth=1.5, label='Critical Temperature')
critical_temp = 2.269

ax.plot(x, magnetizations_32, linewidth=1, label=f'n=32')
ax.plot(x, magnetizations_64, linewidth=1, label=f'n=64')
ax.plot(x, magnetizations_128, linewidth=1, label=f'n=128')
ax.plot(x, magnetizations_256, linewidth=1, label=f'n=256')
ax.plot(x, magnetizations_1024, linewidth=1, label=f'n=1024')

ax.set_xlabel('Temperatures', fontsize=14)
ax.set_ylabel('Magnetization', fontsize=14)
ax.set_title('Magnetization over Temperature with varying lattice sizes', fontsize=16, fontweight='bold')
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend(fontsize=12, loc='upper right')
plt.show()
```
